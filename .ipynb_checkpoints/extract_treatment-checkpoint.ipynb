```python
def init_preprocessing_data(data):
    category = [0, 1, 2, 3, 4]
    radonRisk = list(data.radonRisk.unique())
    t_type = list(data.tumorType.unique())
    stage = list(data.stageAtDiagnosis.unique())
    smoker = list(data.smoker.unique())
    gender = list(data.gender.unique())
    #familialAntecedents = list(data.familialAntecedents.unique())
    #brainMetastasis = list(data.brainMetastasis.unique())

    t_type = [x for x in t_type if str(x) != 'nan']
    stage = [x for x in stage if str(x) != 'nan']
    
    enc_category = MultiLabelBinarizer(classes=category)
    enc_category.fit(category)
    enc_radonRisk = MultiLabelBinarizer(classes=radonRisk)
    enc_radonRisk.fit(radonRisk)
    enc_t_type = MultiLabelBinarizer(classes=t_type)
    enc_t_type.fit(t_type)
    enc_stage = MultiLabelBinarizer(classes=stage)
    enc_stage.fit(stage)
    enc_smoker = MultiLabelBinarizer(classes=smoker)
    enc_smoker.fit(smoker)
    enc_gender = MultiLabelBinarizer(classes=gender)
    enc_gender.fit(gender)
    return enc_category, enc_radonRisk, enc_t_type, enc_stage, enc_smoker, enc_gender

enc_category, enc_radonRisk, enc_t_type, enc_stage, enc_smoker, enc_gender = init_preprocessing_data(patient_impute)
```

```python
def include_category_to_dataset(data, encode, column_name):
    col = to_str(encode.classes, column_name+'-')
    category = pd.DataFrame(columns=col)
    for i in range(data.shape[0]):
        value = data.iloc[i][column_name]
        vector = encode.transform([{value}])
        category.loc[i] = vector[0]
    data = pd.concat([data, category], axis=1, sort=False)
    return data
```

```python
def to_str(list_n, name):
    str_a = ",".join(name + str(x) for x in list_n)
    list_cat = list(str_a.split(","))
    return list_cat
```
