```python
def plot_accuracy(results_th_g1,results_th_g2,results_th_g3, path, metric):
    # set width of bar
    barWidth = 0.1
    # set height of bar (TransD_th85)
    bars1 = list(results_th_g1[metric].values)
    bars2 = list(results_th_g2[metric].values)
    bars3 = list(results_th_g3[metric].values) 
    

    # Set position of bar on X axis
    max_axis = len(bars1) * 0.35
    r1 = np.arange(0, max_axis, 0.35, float)
    #r1 = [0, 0.35, 0.7, 1.05, 1.4]
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    r4 = [x + barWidth for x in r3]

    # Make the plot [color='#13505b'   color='#90bfc9'   color='#2d7f5e']
    plt.bar(r1, bars1, color=mcolors.CSS4_COLORS['rosybrown'], width=barWidth,
            edgecolor='white', label='${\cal{T\_KG}}_{basic}$')
    plt.bar(r2, bars2, color=mcolors.CSS4_COLORS['lightcoral'], width=barWidth,
            edgecolor='white', label='$\cal{T\_KG}$')
    plt.bar(r3, bars3, color=mcolors.CSS4_COLORS['indianred'], width=barWidth,
            edgecolor='white', label='${\cal{T\_KG}}_{random}$')

    # Add xticks on the middle of the group bars
    #plt.ylabel('Accuracy', fontweight='bold')
    plt.ylim(0, 1)
    y1 = np.arange(0, 1.1, 0.1, float)
    plt.yticks(y1)
    model = list(results_th_g1.Model.values)
    plt.xticks([x + 0.1 for x in r1], model, rotation='vertical')
    #plt.xticks([0.1, 0.45, 0.8, 1.15, 1.5], ['TransE', 'TransH', 'RotatE', 'TransD', 'TransR'])
    #plt.title(metric + ' & Threshold '+str(th))
    plt.title(metric)
    """
    for i, v in enumerate(bars1):
        plt.text(r1[i]-0.02, v+0.1, str(v), color='black', fontweight='bold', fontsize='x-small')
    for i, v in enumerate(bars2):
        plt.text(r2[i]-0.02, v+0.1, str(v), color='black', fontweight='bold', fontsize='x-small')
    for i, v in enumerate(bars3):
        plt.text(r3[i]-0.02, v+0.01, str(v), color='black', fontweight='bold', fontsize='x-small')
    """
    # Create legend & Show graphic
    legend = plt.legend(loc='upper right', shadow=False, fontsize='small', ncol=1)
    #plt.savefig(metric + '_Threshold '+str(th)+".png", format='png', bbox_inches='tight', dpi=300)
    #plt.savefig(path + metric + ".png", format='png', bbox_inches='tight', dpi=300)
    plt.savefig(path + metric + ".pdf", format='pdf', bbox_inches='tight')
    plt.show()
```

```python
results_th_g1 = pd.read_csv('G1/results_threshold.csv', sep=';') #results_threshold_G1
results_th_g2 = pd.read_csv('G2/results_threshold.csv', sep=';') #results_threshold_G2
results_th_g3 = pd.read_csv('G3/results_threshold.csv', sep=';') #results_threshold
results_th_g3
```
